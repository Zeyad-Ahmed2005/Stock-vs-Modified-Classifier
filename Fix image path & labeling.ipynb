```python
import pandas as pd
```

```python
df = pd.read_csv("Car_Images_metadata_fixed.csv")
```

```python
def fix_path(p):
    p = p.replace("\\", "/")
    return "Car_Images_Dataset/" + p.split("Car_Images_Dataset/")[-1]

df["image_path"] = df["image_path"].apply(fix_path)

df.to_csv("Car_Images_metadata_fixed.csv", index=False)
```

```python
df["label"] = df["label"].str.strip().str.capitalize()
df["label"] = df["label"].map({
    "Stock": 0,
    "Modified": 1
})

df.to_csv("Car_Images_metadata_fixed.csv", index=False)
```
